## üîß Setup & Tools

```python
def weather_lookup(location):
    if 'monsoon' in location.lower() or 'rain' in location.lower():
        return f"Weather at {location}: Heavy rainfall expected ‚Üí Flood risk HIGH."
    if 'hot' in location.lower() or 'heat' in location.lower():
        return f"Weather at {location}: Heatwave warnings ‚Üí Dehydration risk MEDIUM."
    return f"Weather at {location}: No severe weather reported."

def evacuation_zone(disaster, location=None):
    zones = {
        'flood': "Move to higher ground; avoid low-lying roads.",
        'earthquake': "Drop, cover, hold; move to open area.",
        'fire': "Move crosswind; avoid smoke zones."
    }
    return zones.get(disaster, "No specific evacuation guidance.")

def resource_planner(disaster, severity='moderate'):
    base = {
        'flood': ['Rescue boats', 'Life jackets'],
        'earthquake': ['Medical kits', 'Rescue teams'],
        'fire': ['Water tankers', 'Masks']
    }
    resources = base.get(disaster, ['Generic first aid'])
    if severity == 'high':
        resources += ['Helicopter support']
    return resources

def map_lookup(location):
    return {
        'shelters': [
            {'name': 'Community School', 'distance_km': 2.1},
            {'name': 'Town Hall', 'distance_km': 3.8}
        ],
        'nearest_hospital_km': 4.5
    }
```

## ü§ñ Multi-Agent System Definition

```python
# Agents simulated as simple functions

def weather_agent(location):
    return weather_lookup(location)

def evacuation_agent(disaster, location):
    return evacuation_zone(disaster, location)

def resource_agent(disaster, severity='moderate'):
    return resource_planner(disaster, severity)

def mapping_agent(location):
    return map_lookup(location)
```

## üìä Visualization: Resource Allocation Chart

```python
import matplotlib.pyplot as plt
resources = resource_planner('flood', 'high')
plt.figure(figsize=(8,4))
plt.barh(resources, range(1, len(resources)+1))
plt.title('Resource Allocation for Flood - High Severity')
plt.xlabel('Priority Ranking')
plt.show()
```
